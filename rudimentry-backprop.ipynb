```python
data=pd.read_csv("/kaggle/input/mlware/mnist_train_56x56_grayscale.csv")
Xfull=np.array(data)
Xfull[:,0]=1
X=Xfull[:5000]
print(X)
X=np.array(X,dtype='float64')
m=len(X)
n=len(X[1,:])
classi=pd.read_csv("/kaggle/input/mlware/train.csv")
yfull=np.array(classi)
y=yfull[:5000,1]
print(y)

bias=np.ones([m,1])
ytest=yfull[5000:,1]
Xtest=Xfull[5000:]
Xtest=np.array(Xtest,dtype='float64')
```

```python
np.random.seed(10)
theta1=np.random.rand(n,500)*10
theta2=np.random.rand(501,50)*10
theta3=np.random.rand(51,7)*10
print(theta1,theta2,theta3)
```

```python
def relu(x):
    return np.maximum(0,x)

def sigmoid(x):
    return 1/(1+np.exp(-x))

def softmax(x):
    normx = x - np.max(x)
    exps = np.exp(normx)
    su=np.sum(exps,axis=1)
    b=np.zeros(x.shape)
    for i in range(len(x)):
        b[i,:]=exps[i,:]/su[i]
    return b    

a0=X
print(a0)
z1=(a0@theta1)
print(z1)
a1=sigmoid(z1)
print(a1)
z2=np.append(bias,a1,axis=1)@theta2
a2=sigmoid(z2)
print(a2)
z3=np.append(bias,a2,axis=1)@theta3
print(z3)
##a3=relu(z3)
a3=sigmoid(z3)
print(a3)
print(theta1,theta2,theta3)

```

```python
def forward(X,theta1,theta2,theta3):
    global a1
    global a2
    global a3
    global z1
    global z2
    global z3
    global bias
    a0=X
    z1=(a0@theta1)
    a1=sigmoid(z1)
    z2=np.append(bias,a1,axis=1)@theta2
    a2=sigmoid(z2)
    z3=np.append(bias,a2,axis=1)@theta3
    ##a3=relu(z3)
    a3=sigmoid(z3)
    ##print(a3)
    
print(theta1,theta2,theta3)
def step(x):
    return x>0

def sigmoiddev(x):
    return x*(1-x)
```

```python
d3=a3
for i in range(m):
    d3[i,y[i]]-=1
Del3=np.append(bias,a2,axis=1).T@d3
d2=(d3@(theta3[1:,:].T))*sigmoiddev(a2)
Del2=np.append(bias,a1,axis=1).T@d2
d1=(d2@(theta2[1:,:].T))*sigmoiddev(a1)
Del1=np.append(bias,a0,axis=1).T@d1
print(theta1,theta2,theta3)
```

```python
def backprop(a3,a2,a1,theta3,theta2,y,lambd=0):
    global bias
    global d2
    global d3
    global d1
    global Del1
    global Del2
    global Del3
    global m
    d3=a3
    for i in range(m):
        d3[i,y[i]]-=1
    Del3=np.append(bias,a2,axis=1).T@d3
    d2=(d3@(theta3[1:,:].T))*sigmoiddev(a2)
    Del2=np.append(bias,a1,axis=1).T@d2
    d1=(d2@(theta2[1:,:].T))*sigmoiddev(a1)
    Del1=a0.T@d1
    theta3grad=Del3/m+(lambd/m)*theta3
    theta3grad[0,:]-=(lambd/m)*theta3[0,:]
    theta2grad=Del2/m+(lambd/m)*theta2
    theta2grad[0,:]-=(lambd/m)*theta2[0,:]
    theta1grad=Del1/m+(lambd/m)*theta1
    theta1grad[0,:]-=(lambd/m)*theta1[0,:]
    return [theta3grad,theta2grad,theta1grad]
    
print(theta1,theta2,theta3)
```

```python
alpha=0.05
print(theta1,theta2,theta3)
for i in range(200):
    print(i)
    forward(X,theta1,theta2,theta3)
    theta3grad,theta2grad,theta1grad=backprop(a3,a2,a1,theta3,theta2,y,20)
    theta1=theta1-alpha*theta1grad
    theta2=theta2-alpha*theta2grad
    theta3=theta3-alpha*theta3grad    
    print(theta1,theta2,theta3)
```
